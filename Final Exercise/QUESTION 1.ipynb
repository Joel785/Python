## CALCULATE ROI FOR EACH MOVIE

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### ROI FOR BOLLYWOOD MOVIES

```python
bollywood_df = pd.read_csv("cleaned_bollywood.csv")
bollywood_df.head()
```

```python
"""
Calculating Profit by BoxOffice Collections - Budget and then calculation ROI by Profit/Budget
"""
bollywood_df['Profit (Crores)']=bollywood_df['BoxOffice (Crores)']-bollywood_df['Budget (Crores)']
bollywood_df['ROI']=bollywood_df['Profit (Crores)']/bollywood_df['Budget (Crores)']
bollywood_df.head(10)
```

```python
"""
Sorting the data by ROI  to find the films in the order ROI
"""
Top_3_Bolly=bollywood_df.sort_values(by='ROI', ascending= False)
Top_3_Bolly.head(3)
```

```python
Top_3_Bolly=Top_3_Bolly[['Title','ROI']]
Result_Bollywood = Top_3_Bolly.head(3)
Result_Bollywood
```

## GRAPHS FOR BOLLYWOOD MOVIES

```python
plt.figure(figsize=(10, 5))
plt.bar(bollywood_df['Title'], bollywood_df['ROI'], color='skyblue')
plt.title("Bollywood Movies by ROI")
plt.xlabel("Movie Title")
plt.ylabel("ROI")
plt.xticks(rotation=90, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(Result_Bollywood['Title'], Result_Bollywood['ROI'], color='skyblue')
plt.title("Top 3 Bollywood Movies by ROI")
plt.xlabel("Movie Title")
plt.ylabel("ROI")
plt.xticks(rotation=45)
plt.show()
```
