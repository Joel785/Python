```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
"""
Combining all the datasets
"""
df_bollywood= pd.read_csv('cleaned_bollywood.csv')
df_tollywood= pd.read_csv('cleaned_tollywood.csv')
df_indian= pd.read_csv('cleaned_indian.csv')
df_combined = pd.concat([df_bollywood, df_tollywood, df_indian])
```

```python
df_combined['BoxOffice (Crores)'].mean()
```

```python
"""
Graph showing distribution of box office collections. 
"""
plt.figure(figsize=(10, 5))
plt.hist(df_combined['BoxOffice (Crores)'], bins=10, edgecolor='black', color='skyblue')
plt.title('Box Office Distribution')
plt.xlabel('Box Office (Crores)')
plt.ylabel('Number of Movies')
plt.show()
```

```python
"""
The box whisker graph for the combined data showing distribution of box office collections. 
"""
plt.boxplot(df_combined['BoxOffice (Crores)'])
plt.title('BoxOffice (Crores)')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.savefig('All Indian movies Genre avg box office.png')
```

## INFERENCE

### 1. Box Office Distribution Plot

- Most movies have box office collections in the lower end, likely below **200 crores**.
- The long **right tail** indicates that a few movies have exceptionally high collections, suggesting a **positive skew**.

---

### 2. Box Office (Crores) – Outlier Detection (Box Plot)

- The **box plot** confirms that most data lies well below the upper whisker, likely below **200–250 crores**.
- There are clear **outliers** that fall far beyond the upper range of the normal distribution.
