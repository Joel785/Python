```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
"""
Combining all the three industries
"""
df_bollywood= pd.read_csv('cleaned_bollywood.csv')
df_tollywood= pd.read_csv('cleaned_tollywood.csv')
df_indian= pd.read_csv('cleaned_indian.csv')
df_combined = pd.concat([df_bollywood, df_tollywood, df_indian])
```

## AVERAGE CALCULATION FOR ALL THE COMBINED GENRE

```python
combined_avg_genre = round(df_combined.groupby('Genre')['BoxOffice (Crores)'].mean(), 2)
combined_avg_genre 
```

```python
combined_avg_genre.sort_values(ascending=False).plot.barh(
    title='All Indian movies Genre avg box office', figsize=(10, 6), color='skyblue'
)
plt.xlabel('Box Office (Crores)')
plt.ylabel('Genre')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
plt.savefig('All Indian movies Genre avg box office.png')
```

## AVERAGE CALCULATION FOR ALL THE COMBINED GENRE

```python
bolly_avg_genre = round(df_bollywood[df_bollywood['Language'] == 'Hindi'].groupby('Genre')['BoxOffice (Crores)'].mean(), 2)
```

```python
bolly_avg_genre_sorted = bolly_avg_genre.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(bolly_avg_genre_sorted.index, bolly_avg_genre_sorted.values, color='lightcoral')
plt.title('Bollywood Genre Avg Box Office')
plt.xlabel('Average Box Office (Crores)')
plt.ylabel('Genre')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()


```

```python
plt.savefig('Bollywood Genre Avg Box Office.png')
```

**Low Performing Genres**:
- Genres such as Mystery,Horror, and Romance are performing below average in the combined dataset.
